```python
# Missing data : Data that contains empty cells (NaN.)
def function_showmissingdata():
  emptyline=df.isnull().any(axis=1)
  missingdata=df[emptyline]
  print("Here is how many data missing we have ", df.isnull().any(axis = 1).sum())
  print(missingdata)
```

**It seems like Message column has no data. In the following cells, we will investigate if there is more columns that has also no data so we will be aware of** 

```python
# defining a function that can take care of each columns and check if that columns contains data or not. 
def detectingmissingdatainColumn(x):
  cond=df[x].isnull()
  txt=""
  if len(df[cond])==0:
    return "The Coulmn: ", x , " field contains data"
  else:
    return "The Column: ", x , " has missing data: ", len(df[cond])
    
```

```python
#Answer for 1: 
# a) let us pick good or Acceptable AQI for the 5 cities in Brorough
import re
cond1=((df['Data Value']<=100) 
        & 
       (df['Geo Type Name']=='Borough')
        )
table=df[cond1]
regex=re.compile(r'[0-9]{4}')
year_list=[]
for item in table['Start_Date']:
  item=str(item)
  match=regex.finditer(item)
  for me in match:
    year_list.append(me.group(0))
data={
    'Borough':table['Geo Place Name'],
    'Value':table['Data Value'],
    'Year':year_list
      
}
data1=pd.DataFrame(data)
first1= data1.pivot_table(values="Value", index=["Borough"],
           columns="Year", aggfunc="mean", margins=True,fill_value=0,
            margins_name='Total')
print("Showing good or satisfactory AQI within Borough for years")
first1
```

```python
pivot = data1.pivot_table(index=['Borough'], values=['Value'], aggfunc='mean').plot()

plt.show()
```

**We carfully can say that Manhattan seems to lead other cities in high AQI index of air quality than other cities, while, Staten Island seems to have nice air quality comparing to other cities. However, we can not conclude anything yet**

```python
# a) continue dealing with Moderate or more bad AQI 
cond2=((df['Data Value']>100)
        &
        (df['Geo Type Name']=='Borough')
          )
table1=df[cond2]
regex=re.compile(r'[0-9]{4}')
year_list=[]
for item in table1['Start_Date']:
  item=str(item)
  match=regex.finditer(item)
  for me in match:
    year_list.append(me.group(0))
data={
      'Borough':table1['Geo Place Name'],
      'Value':table1['Data Value'],
      'Year':year_list
        
  }
data2=pd.DataFrame(data)
first2= data2.pivot_table(values="Value", index=["Borough"],
            columns="Year", aggfunc="mean", margins=True,fill_value=0,
              margins_name='Total')
print("Showing Bad or worst AQI within Borough for years")

first2
```

```python
pivot = data2.pivot_table(index=['Borough'], values=['Value'], aggfunc='mean').plot()

plt.show()
```

```python
# Another way to see that by using pie chart
plot_data={
      'Value':[156.703688,102.650000,137.839116]
}

mydata=pd.DataFrame(plot_data,index=['Bronx','Brooklyn','Manhattan'])

plot = mydata.plot.pie(y='Value', figsize=(11, 11))
```

```python
# Answer:2 For no Bourough
# a) let us pick good or Acceptable AQI for no Brorough

cond1=((df['Data Value']<100) 
        & 
       (df['Geo Type Name']!='Borough')
        )
table=df[cond1]
regex=re.compile(r'[0-9]{4}')
year_list=[]
for item in table['Start_Date']:
  item=str(item)
  match=regex.finditer(item)
  for me in match:
    year_list.append(me.group(0))
data={
    'other':table['Geo Type Name'],
    'Value':table['Data Value'],
    'Year':year_list
}
data1=pd.DataFrame(data)
first1= data1.pivot_table(values="Value", index=["other"],
           columns="Year", aggfunc="mean", margins=True,fill_value=0,
            margins_name='Total')
print("Showing good or satisfactory AQI within no Borough for years")
first1
```

```python
#Question 6
my_year=[]
regex=re.compile(r'([0-9]{4})')
for year in df['Start_Date']:
  item=str(year)
  matches=regex.finditer(item)
  for match in matches:
    my_year.append(match.group(1))

data = {
       'Year':my_year,
        'collection':df['Unique ID'],
     'Average':df['Data Value']
}
mytable=pd.DataFrame.from_dict(data)
mytable_pivot=pd.pivot_table(mytable, index=['Year'], aggfunc={'collection':'count', 'Average':'mean'}
                )

mytable_pivot=mytable_pivot.sort_values('Average',axis=0,ascending=False)
mytable_pivot
```
